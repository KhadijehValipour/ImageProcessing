```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.zeros((300 , 300) , dtype=np.uint8)   # Construction of the first layer

r[0:150 , 0:150] = 255
r[150:300 , 150:300] =255

plt.imshow(r , cmap='gray')
```

```python
g = np.zeros((300,300),dtype=np.uint8)  # Construction of the second layer

g[0:150 ,150:300] = 255
g[150:300 , 150:300] = 255

plt.imshow(g , cmap='gray')
```

```python
b = np.zeros((300,300) , dtype=np.uint8)  # Construction of the third layer

b[150:300 , 0:150] =255

plt.imshow(b , cmap='gray')
```

```python
microsoft_logo = cv2.merge((r , g , b))

plt.imshow(microsoft_logo)
plt.show()
```
